```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load 
fire_csv = "Datasets/Fires.csv"
weather_csv = "Weather_Data.csv"

#List of columns in the csv to use
columns = ['LATITUDE',
           'LONGITUDE',
           'STATE',
           'STAT_CAUSE_DESCR',
           'DISCOVERY_DOY',
           'CONT_DOY',
           'FIRE_YEAR',
           'FIRE_SIZE',
           'FIRE_SIZE_CLASS']
# Read in the wildfire csv, only first 1000 rows for testing (nrows=1000)
fires_df = pd.read_csv(fire_csv, nrows=100, usecols=columns)
weather = pd.read_csv(weather_csv, nrows=1000)
fires_df.head()
```

```python
weather.info()
```

```python
def dist(lat1, long1, lat2, long2):
    return np.abs(((float(lat1)-float(lat2))**2+(float(long1)-float(long2))**2)**0.5)

def find_station(lat, long):
    distances = weather.apply(
        lambda row: dist(lat, long, row['Latitude'], row['Longitude']), 
        axis=1)
    return weather.loc[distances.idxmin(), 'ID']

fires_df.apply(
    lambda row: find_station(row['LATITUDE'], row['LONGITUDE']), 
    axis=1)
```
