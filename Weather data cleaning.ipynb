```python
import pandas as pd
import os,sys,glob
import numpy as np
from citipy import citipy
```

```python
#Get path to pull the city data

pd_city_info = pd.read_csv('city_info.csv')
pd_city_info.head()
```

```python
#drop duplicate values from Name column
pd_city_final =pd_city_info.drop_duplicates(subset='Name', keep='first', inplace=False).reset_index(drop=True)

#renamed the unnamed:) column and dropped it
pd_city_final.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True)
pd_city_final.drop(columns='a',inplace = True)

#drop multiple columns
pd_city_final.drop(["Stn.Name","Stn.stDate","Stn.edDate"],axis=1,inplace = True)
pd_city_final.head()
```

```python
#Reading the csv files and concatinating the csvs into one datframe

path = ('2000-2015')
filenames = glob.glob(path + "/*.csv")

df = pd.concat([pd.read_csv(fp).assign(New=os.path.basename(fp)) for fp in filenames])
df = df.rename(columns={'New':'ID'})
df['ID'] = df['ID'].str.split('.').str[0]
df.head()
```

```python
# Cleaning the dataframe

df=df.dropna()
df['Date'] = pd.to_datetime(df['Date'])

df.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True)
df.drop(columns='a',inplace = True)

df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month

df.head()
```

```python
#Getting rows for years from 2000-2015
array =np.arange(2000,2016,1)
df_final = df.loc[df['Year'].isin(array)]
df_final.reset_index(drop=True, inplace=True) 
df_final.head()
```

```python
#df_final.to_csv(r'Weather_Data_Final.csv')
```

```python
df_merge = pd.merge(df_final,pd_city_final)
df_merge.head()
```

```python
df_merge.info()
```

```python
#Declare as numeric
df_merge['tmax'] = df_merge['tmax'].fillna(0).astype(int)
df_merge['tmin'] = df_merge['tmin'].fillna(0).astype(int)
df_merge['prcp'] = df_merge['prcp'].fillna(0).astype(int)
df_merge.info()
```

```python
df_mean= pd.DataFrame(df_merge.groupby(['ID','Year','Month']).
                      agg(Mean_Tmax=pd.NamedAgg('tmax','mean'),
                          Mean_Tmin=pd.NamedAgg('tmin','mean'),
                          Mean_Prcp=pd.NamedAgg('prcp','mean')).reset_index())
df_mean.head()
```

```python
df_final_mean = pd.merge(df_mean,pd_city_final)
df_final_mean.head()
```
