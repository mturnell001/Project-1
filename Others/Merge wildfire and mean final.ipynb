```python
import pandas as pd
import os,sys,glob
import numpy as np
from citipy import citipy

```

```python
wildfire_data = pd.read_csv('wildfires 2000-2015.csv')
wildfire_data = wildfire_data.head(1000)
wildfire_data.head()
```

```python
df_final_mean =pd.read_csv('Mean_Weather_Data_Final.csv')
df_final_mean.head(100)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = wildfire_data['LATITUDE']
lngs = wildfire_data['LONGITUDE']
lat_lngs = zip(lats, lngs)

wildfire_data['CITY']=''
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    cities.append(city.title())
    
  
wildfire_data['CITY']=cities
# Print the city count to confirm sufficient count
wildfire_data.head()
```
